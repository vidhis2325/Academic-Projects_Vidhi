# MODEL 1 ols

```python
import pandas as pd
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Assuming your DataFrame is already loaded in `df`

# Convert all relevant columns to numeric, coercing errors to NaN
cols_to_convert = ['lnRE', 'lngdp', 'lnepu', 'lnco2pc', 'lnto', 'lndcps', 'lnrnd', 'lnpopg', 'lnene']
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Drop rows with any NaN values
df.dropna(inplace=True)

# Define the dependent and independent variables
y_col = 'lnRE'
X_cols = ['lngdp', 'lnepu', 'lnco2pc', 'lnto', 'lndcps', 'lnrnd', 'lnpopg', 'lnene']

# Define the dependent variable (y) and independent variables (X)
y = df[y_col]
X = df[X_cols]

# Add a constant to the independent variables matrix for the intercept
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Get the model summary
print(model.summary())

# Predict the values
y_pred = model.predict(X)

# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse}")

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Display the VIF data
print("\nVariance Inflation Factors (VIF):")
print(vif_data)

# Calculate the mean VIF (excluding the constant)
mean_vif = vif_data.loc[vif_data['Feature'] != 'const', 'VIF'].mean()
print(f"\nMean VIF (excluding the constant): {mean_vif}")
```

# Endogeneity and IV-GMM
Endogeneity can occur if one or more of the explanatory variables is correlated with the error term
 In this case, Economic Policy Uncertainty (EPU) might cause endogeneity issues if:
        
    EPU is correlated with the error term: If economic policies affecting EPU also directly or indirectly affect RE growth, EPU could be endogenous. This means changes in EPU might be driven by factors that also impact RE growth, leading to a correlation between epu and error term    

```python
import pandas as pd
import numpy as np
from linearmodels.iv import IVGMM
from statsmodels.tools.tools import add_constant
from scipy import stats

# Load your dataset

# Define the variables
y_col = 'lnRE'
X_cols = ['lngdp', 'lnepu', 'lnco2pc', 'lnto', 'lndcps', 'lnrnd', 'lnpopg', 'lnene']

# Create lagged variables for lnepu
df['lnepu_lag1'] = df['lnepu'].shift(1)
df['lnepu_lag2'] = df['lnepu'].shift(2)

# Drop rows with NaN values due to lagging
df = df.dropna()

# Define the endogenous variable and instruments
endog = 'lnepu'
instruments = ['lnepu_lag1', 'lnepu_lag2']

# Define the dependent variable and exogenous variables
y = df[y_col]
X = df[X_cols]

# Add a constant term for the exogenous variables
X = add_constant(X)

# Specify the model
model = IVGMM.from_formula(
    f'{y_col} ~ 1 + lngdp + lnco2pc + lnto + lndcps + lnrnd + lnpopg + lnene + [lnepu ~ lnepu_lag1 + lnepu_lag2]',
    df
)

# Fit the model
results = model.fit()

# Print the results
print(results.summary)

# Hansen J test for instrument validity
# Extract residuals
residuals = results.resids

# Compute the Hansen J statistic
instrument_matrix = df[instruments].values
num_instruments = instrument_matrix.shape[1]
num_observations = instrument_matrix.shape[0]

# Compute the Hansen J statistic
moment_conditions = np.dot(instrument_matrix.T, residuals)
J_statistic = np.dot(moment_conditions.T, moment_conditions) / num_observations
p_value = 1 - stats.chi2.cdf(J_statistic, num_instruments - 1)

print(f'Hansen J Statistic: {J_statistic}')
print(f'P-Value: {p_value}')
```

# Granger Casuality

```python
from statsmodels.tsa.stattools import grangercausalitytests
import pandas as pd
df.dropna(inplace=True)
pairs = [('lnepu', 'lnRE'), ('lnRE', 'lnepu'), ('lnepu', 'lnhydro'),
         ('lnhydro', 'lnepu'), ('lnepu', 'lnothers'), ('lnothers', 'lnepu')]

null_hypotheses = []
F_statistics = []
p_values = []

for pair in pairs:
    results = grangercausalitytests(df[[pair[0], pair[1]]], maxlag=2, verbose=False)
    null_hypothesis = f"{pair[0]} does not Granger-cause {pair[1]}"
    F_statistic = results[2][0]['params_ftest'][0]
    p_value = results[2][0]['params_ftest'][1]
    null_hypotheses.append(null_hypothesis)
    F_statistics.append(F_statistic)
    p_values.append(p_value)

table_data = {
    "Null hypothesis": null_hypotheses,
    "F-statistic": F_statistics,
    "p-value": p_values
}
table_df = pd.DataFrame(table_data)

print("Causality analysis Table:")
print(table_df.to_string(index=False))
```
